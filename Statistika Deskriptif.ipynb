a. Distribusi Uniform

---



---



a)	Carilah nilai a dan b terlebih dahulu

```python
ph = ([3.5] , [1.5], [8.2], [2.6], [4.5], [9.1], [2.7], [6.8], [4.3], [3.5],
      [3.6], [7.4], [7.3], [8.9], [7.1], [2.1], [9.5], [2.8], [2.3], [8.4],
      [8.6], [12], [2.1], [3.5], [1.5], [6.7], [5.5], [6.7], [2.6], [5.5],
      [6.6], [6.9], [8.3], [8.9], [7.5], [4.3], [9.7], [2.6], [3.3], [8.8])

a = float('inf')
b = float('-inf')

for sublist in ph:
    a_sublist = min(sublist)
    b_sublist = max(sublist)
    a = min(a, a_sublist)
    b = max(b, b_sublist)

print("nilai terkecil:", a)
print("nilai terbesar:", b)
```

b)	Gunakan persamaan mean dan standard deviation sebagai bekal untuk mendapatkan nilai a dan b sesuai dengan sifat dasar statistika dalam distribusi uniform berikut

```python
import math

mean = (a + b) / 2
std_dev = (b - a) / math.sqrt(12)

print("mean", mean)
print("standar deviasi", std_dev)
```

c)	Selanjutnya hitung probabilitas distribusi uniform berdasarkan nilai a dan b yang sudah didapatkan menggunakan persamaan berikut

```python
x = 5
probabilitas = (x - a) / (x - b)
presentase = probabilitas*100

print("probabilitas", probabilitas)
print("presentase", presentase)
```

d)	Visualisasikan dataset menggunakan histogram, periksalah apakah dataset berdistribusi uniform atau tidak. Jelaskan beserta alasannya.

```python
import matplotlib.pyplot as plt
import numpy as np

array = np.array(ph)
prob = -0.5
if prob == -0.5:
    print('Data berdistribusi uniform')
else:
    print('Tidak berdistribusi uniform')

plt.hist(array, bins=20, density=True, alpha=0.7, color='pink')
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.title('Histogram Data')
a = 2
b = 6
plt.axvline(x=a, color='r', linestyle='--', label=f'a = {a:.2f}')
plt.axvline(x=b, color='g', linestyle='--', label=f'b = {b:.2f}')
plt.axvline(x=5, color='y', linestyle='--', label='x = 5')
plt.legend()
plt.show()
```

b. Distribusi Normal

---



---



```python
import pandas as pd
normal_distribution = pd.read_csv("/content/normal_distribution.csv", sep=',')
normal_distribution
```

a)	Hitunglah persamaan f(x) dari nilai x di atas berdasarkan persamaan dalam distribusi normal dengan menggunakan kode script pemrograman Python secara scratch.

```python
import numpy as np

mean = np.mean(normal_distribution['x'])
std = np.std(normal_distribution['x'])

print("mean", mean)
print("standar deviasi", std)
```

```python
import numpy as np

normal_distribution['x']=normal_distribution['x'].astype(float)
normal_distribution['f(x)'] = 1 / (std * np.sqrt(2 * np.pi)) * np.exp(-(normal_distribution - mean)**2 / (2 * std**2))
print(normal_distribution)
```

b)	Visualisasikan antara nilai x dan f(x) menggunakan diagram garis, periksalah apakah dataset berdistribusi normal atau tidak. Jelaskan beserta alasannya.

```python
import matplotlib.pyplot as plt
from scipy.stats import shapiro

statistic, p_value = shapiro(normal_distribution ['x'])
alpha = 0.05
if p_value > alpha :
    print('Data Berdistribusi Normal')
else:
    print('Data tidak Berditribusi Normal')

plt.plot(normal_distribution['x'], normal_distribution['f(x)'])
plt.title("Normal Distribution")
plt.xlabel("Nilai x")
plt.ylabel("Nilai f(x)")
plt.grid(True)
plt.show()
```

c. Skewness

---



---



```python
import pandas as pd
skew = pd.read_csv("/content/skewness.csv", sep=';')
skew
```

a)	Hitunglah persamaan f(x) dari nilai x di atas berdasarkan persamaan dalam distribusi normal dengan menggunakan kode script pemrograman Python secara scratch.

```python
kolomx = skew['x']
xtitik = kolomx.str.replace(',', '.')
xfloat = xtitik.astype(float)
mean = np.mean(xfloat)
std = np.std(xfloat)
print("Mean data distribusi normal adalah", mean)
print("Standar Deviasi distribusi normal adalah", std)
skew['f(x)'] = 1 / (std*np.sqrt(2*np.pi)) * np.exp(-(xfloat - mean)**2 / (2 * std**2))
skew
```

b)	Visualisasikan antara nilai x dan f(x) menggunakan diagram garis, periksalah apakah dataset berdistribusi normal atau tidak. Jelaskan beserta alasannya.

```python
import matplotlib.pyplot as plt
from scipy.stats import shapiro

p_value, statistic = shapiro(xfloat)
alpha = 0.05
if p_value > alpha:
  print('Data berdistribusi normal')
else:
  print('Data tidak berdistribusi normal')

plt.plot(xfloat, skew['f(x)'])
plt.title('Normal Distribution')
plt.xlabel('X')
plt.ylabel('f(x)')
plt.grid(True)
plt.show
```

c)	Selanjutnya, gunakan persamaan skewness sebagai berikut (ke Python):

```python
mean = np.mean(xfloat)
median = np.median(xfloat)
std = np.std(xfloat)
print("Mean", mean)
print("Median", median)
print("Standar Deviasi", std)

skewness = 3*(mean-median) / std
print("Skewness", skewness)

if skewness > 0:
  print("Distribusi dataa condong ke arah positif")
  print("sebesar", skewness)
elif skewness < 0:
  print("Distribusi data contong ke arah negatif")
  print("sebesar", skewness)
else:
  print("Distribusi data normal")
```

d. Kurtosis

---



---



```python
import pandas as pd
kurtosis = pd.read_csv("/content/kurtosis.csv", sep=';')
kurtosis
```

a)	Hitunglah persamaan f(x) dari nilai x di atas berdasarkan persamaan dalam distribusi normal dengan menggunakan kode script pemrograman Python secara scratch.

```python
kolomx = kurtosis['x']
xtitik = kolomx.str.replace(',', '.')
xfloat = xtitik.astype(float)
mean = np.mean(xfloat)
std = np.std(xfloat)
print("Mean data distribusi normal adalah", mean)
print("Standar Deviasi distribusi normal adalah", std)
kurtosis['f(x)'] = 1 / (std*np.sqrt(2*np.pi)) * np.exp(-(xfloat - mean)**2 / (2 * std**2))
kurtosis
```

b)	Visualisasikan antara nilai x dan f(x) menggunakan diagram garis, periksalah apakah dataset berdistribusi normal atau tidak. Jelaskan beserta alasannya.

```python
import matplotlib.pyplot as plt
from scipy.stats import shapiro

p_value, statistic = shapiro(xfloat)
alpha = 0.05
if p_value > alpha:
  print('Data berdistribusi normal')
else:
  print('Data tidak berdistribusi normal')

plt.plot(xfloat, kurtosis['f(x)'])
plt.title('Normal Distribution')
plt.xlabel('X')
plt.ylabel('f(x)')
plt.grid(True)
plt.show
```

c)	Distribusi data mengarah ke platykurtic, mesokurtic, atau leptokurtic? Jelaskan beserta alasannya.

```python
data_kurto = len(xfloat)
print('Banyaknya data dari kurtosis sebesar', data_kurto)
xbar = sum(xfloat) / (data_kurto)
print('x bar', xbar)
n = len(xfloat)
s4 = np.std(xfloat)
a4 = (1/n) * np.sum((xfloat  - xbar)**4) / (s4**4)
print('Data Kurtosis', a4)

if a4 > 3:
  print('Leptokurtic', 'dengan nilai kurtosis sebesar', a4)
elif a4 < 3:
  print('Platykurtic', 'dengan nilai kurtosis sebesar', a4)
else:
  print('Mesokurtic', 'dengan nilai kurtosis sebesar', a4)
```

# 2. Grouping Dataset: Pivoting Table

```python
import pandas as pd
dbp = pd.read_csv("/content/data_bahan_pokok.csv", sep=';')
dbp
```

Pivoting Tabel Berdasarkan Jumlah Pembelian Bahan Pokok (kg) per Kuartal

---



---



```python
jumlah = dbp.pivot_table(index=['BARANG', 'KUARTAL'],
                         columns = 'TAHUN', values = 'JUMLAH',
                         aggfunc = 'sum', fill_value = 0)
jumlah['GRAND TOTAL'] = jumlah.sum(axis=1)
jumlah
```

Pivoting Tabel Berdasarkan Total Pembelian Bahan Pokok (Rupiah) per Kuartal

---



---



```python
dbp['TOTAL'] = dbp['TOTAL'].str.replace('IDR', '').str.replace('.','').str.replace(',','.').astype(float)
dbp

total = dbp.pivot_table(index=['BARANG', 'KUARTAL'], columns='TAHUN', values='TOTAL',
                              aggfunc='sum', fill_value=0)

total ['Grand Total']=total.sum(axis=1)
total
```
